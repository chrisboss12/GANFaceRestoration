## Model Design

```python
def define_generator(in_shape=(256, 256, 3)):
    # Encoder
    generator_input = Input(in_shape)

    x = Conv2D(64, 5, padding='same', input_shape=(256, 256, 3))(generator_input)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, 3, padding='same', strides=(2, 2))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, 3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same', strides=(2, 2))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Dilation layers
    x = Conv2D(256, 3, padding='same', dilation_rate=(2, 2))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same', dilation_rate=(4, 4))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same', dilation_rate=(8, 8))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same', dilation_rate=(16, 16))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(256, 3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    # Decoder
    x = Conv2DTranspose(128, 4, padding='same', strides=(2, 2))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(128, 3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2DTranspose(64, 4, padding='same', strides=(2, 2))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(32, 3, padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    x = Conv2D(3, 3, padding='same')(x)
    generator_output = Activation('sigmoid')(x)

    model = Model(generator_input, outputs=generator_output, name='Generator')
    return model
```

```python
def define_discriminator(in_shape=(256,256,3)):
    global_disc_input=Input(in_shape)

    x=Conv2D(64, 5, padding='same', input_shape = (256,256,3),strides=(2,2))(global_disc_input)
    x=BatchNormalization()(x)
    #x=Activation('relu')(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(128, 5, padding='same',strides=(2,2))(x)
    x=BatchNormalization()(x)
    #x=Activation('relu')(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(256, 5, padding='same',strides=(2,2))(x)
    x=BatchNormalization()(x)
    #x=Activation('relu')(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(512, 5, padding='same',strides=(2,2))(x)
    x=BatchNormalization()(x)
    #x=Activation('relu')(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=Conv2D(512, 5, padding='same',strides=(2,2))(x)
    x=BatchNormalization()(x)
    #x=Activation('relu')(x)
    x=LeakyReLU(alpha=0.2)(x)

    x=layers.Flatten()(x)
    x=Dense(512,activation='relu')(x)
    global_disc_output=Dense(1,activation='sigmoid')(x)

    model=Model(global_disc_input,outputs=global_disc_output, name='Global_disc')
    opt = Adam(learning_rate=0.0002, beta_1=0.5)
    model.compile(loss='binary_crossentropy', optimizer=opt)
    model.summary()
    return model


d_model = define_discriminator(in_shape=(256,256,3))
g_model = define_generator(in_shape=(256,256,3))
```

```python
def define_gan(g_model, d_model, in_shape=(256,256,3)):
    # make weights in the discriminator not trainable
    d_model.trainable = False
    # define the source image
    in_src = Input(shape=in_shape)
    # connect the source image to the generator input
    gen_out = g_model(in_src)
    # connect the source input and generator output to the discriminator input
    dis_out = d_model(gen_out)
    # src image as input, generated image and classification output
    model = Model(in_src, [dis_out, gen_out])
    # compile model
    opt = Adam(learning_rate=0.0002, beta_1=0.5)
    model.compile(loss=['binary_crossentropy', 'mse'], optimizer=opt, loss_weights=[1,100])
    return model

gan_model=define_gan(g_model, d_model, in_shape=(256,256,3))
```

```python
# select a batch of random samples, returns images and target

def generate_real_samples(dataset, n_samples):
    trainA=dataset
    
    # choose random instances
    ix = randint(0, trainA.shape[0], n_samples)
    
    # retrieve selected images
    X1= trainA[ix]
 #  print(X1.shape)
    
    # generate 'real' class labels (1)
    y = ones((n_samples, 1))
    # print(y.shape)
    
    return X1, y,ix


# generate a batch of images, returns images and targets
def generate_fake_samples(g_model, samples):
    # generate fake instance
    X = g_model.predict(samples)
    # create 'fake' class labels (0)
    y = zeros((len(X), 1))
    return X, y
```

```python
def train(d_model, g_model, gan_model, dataset_real, dataset_damaged, n_epochs=100, n_batch=32, n_batch_test=10):
    # unpack dataset
    trainA = dataset_real
    # calculate the number of batches per training epoch
    bat_per_epo = int(len(trainA) / n_batch)
    # calculate the number of training iterations
    n_steps = bat_per_epo * n_epochs
    # manually enumerate epochs

    dlossreal = []
    dlossfake = []
    glossbce = []
    glossmae = []
    steps = []
    
    for k in range(1, n_epochs + 1):
        for i in range(bat_per_epo):
            # select a batch of real samples
            X_real, y_real, _ = generate_real_samples(dataset_real, n_batch)
            
            # generate random indices for damaged images
            image_indices = np.random.randint(0, len(dataset_damaged), n_batch)
            
            # generate a batch of fake samples
            X_fake, y_fake = generate_fake_samples(g_model, dataset_damaged[image_indices])
            
            if i == 124:
                plt.figure(figsize=(20, 6))
                for m in range(5):
                    # Display original
                    ax = plt.subplot(3, 5, m + 1)
                    plt.imshow(X_fake[m][:, :, ::-1])
                    ax.get_xaxis().set_visible(False)
                    ax.get_yaxis().set_visible(False)

                    ax = plt.subplot(3, 5, m + 1 + 5)
                    plt.imshow(X_real[m][:, :, ::-1])
                    ax.get_xaxis().set_visible(False)
                    ax.get_yaxis().set_visible(False)

                    ax = plt.subplot(3, 5, m + 1 + 10)
                    plt.imshow(dataset_damaged[image_indices][m][:, :, ::-1])
                    ax.get_xaxis().set_visible(False)
                    ax.get_yaxis().set_visible(False)

                plt.show()

            # update discriminator for real samples
            d_loss_real = d_model.train_on_batch(X_real, y_real)
            # update discriminator for generated samples
            d_loss_fake = d_model.train_on_batch(X_fake, y_fake)
            # update the generator
            gloss_all, g_loss_BCE, g_loss_mae = gan_model.train_on_batch(dataset_damaged[image_indices], [y_real, X_real])

            # summarize performance
            if i == 124:
                print('epochs[%d],batch[%d], d_real[%.5f] d_fake[%.5f] g_BCE[%.5f] g_mae[%.5f]' % (k, i + 1, d_loss_real, d_loss_fake, g_loss_BCE, g_loss_mae))

            dlossreal.append(d_loss_real)
            dlossfake.append(d_loss_fake)
            glossbce.append(g_loss_BCE)
            glossmae.append(g_loss_mae)
            steps.append(k * i)
    
    plt.plot(steps, dlossreal)
    plt.plot(steps, dlossfake)
    plt.title('Discriminator loss')
    plt.ylabel('loss')
    plt.xlabel('batches')
    plt.legend(['Discriminator real loss', 'Discriminator fake loss'], loc='upper right')
    plt.show()

    plt.plot(steps, glossbce)
    plt.plot(steps, glossmae)
    plt.title('Generator loss')
    plt.ylabel('loss')
    plt.xlabel('batches')
    plt.legend(['GAN BCE train loss', 'Generator MAE train loss'], loc='upper right')
    plt.show()
```
